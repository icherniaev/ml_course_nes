## 2. Визуализации и matplotlib

При работе с данными часто неудобно делать какие-то выводы, если смотреть на таблицу и числа в частности, поэтому важно уметь визуализировать данные. В этом разделе мы этим и займёмся.

- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторое величины;
- scatter — график, показывающий взаимосвязь двух величин;
- bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.

В этом задании вы попробуете построить каждый из них. Не менее важно усвоить базовые принципы визуализаций:
- на графиках должны быть подписаны оси;
- у визуализации должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах).

Изображайте на двух соседних графиках. Для этого может быть полезен subplot.

```python
df.groupby('order_id', as_index=False).agg({'item_price': 'sum'})
```

```python
fig, ax = plt.subplots(1,2, figsize=(20,10))

fig.tight_layout()
# plot the first one
ax[1].hist(x='item_price', data=df.groupby('order_id', as_index=False).agg({'item_price': 'sum'}),
        bins=100)
ax[1].set_title('Purchase sums histogram')
ax[1].set_xlabel('Total purchase price')
ax[1].set_ylabel('Number of purchases at a given price')

# plot the second one

ax[0].barh(y='item_name', width='one_item_price', data=df.groupby('item_name', as_index=False).agg({'one_item_price': 'mean'}) \
    .sort_values('one_item_price', ascending=False))

ax[0].set_title('Histogram of average prices per category')
ax[0].set_xlabel('Average price')
ax[0].set_ylabel('Category');
```

```python
fig, ax = plt.subplots(figsize=(12,4))

plt.plot(df.groupby('date', as_index=False).agg({'item_price': 'sum'}).sort_values('date')['date'],
    df.groupby('date', as_index=False).agg({'item_price': 'sum'}).sort_values('date')['item_price'])

plt.xticks(rotation = 45);
ax.set_title('Total sum of purchase / Date dependence')
ax.set_xlabel('Date')
ax.set_ylabel('Total sum of purchase in $')
```

#### 3. [1 балл] Постройте средних сумм покупок по дням недели (bar plot).

```python
# convert date column to datetime64 type
df['date'] = df['date'].astype('datetime64')
#create column for weekday
df['weekday'] = df.date.dt.weekday
```

```python
# dataframe with total money spend during a weekday
sums = df.groupby('weekday', as_index=False).agg({'item_price': 'sum'})
# dataframe with total number of orders during a weekday
counts = df.drop_duplicates(subset=['order_id']).groupby('weekday', as_index=False).agg({'order_id': 'count'})
# merged dataframe
tmp = sums.merge(counts, on='weekday')
tmp['spent_per_weekday'] = tmp['item_price'] / tmp['order_id']
```

```python
fig, ax = plt.subplots(figsize=(12,4))

plt.bar(x='weekday',
        height='spent_per_weekday',
        data = tmp)
plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
ax.set_title('Average purchase by day of week')
ax.set_xlabel('Weekday')
ax.set_ylabel('Sum of purchase in a weekday');
```
